```python
df = pd.read_csv('error01051.txt',names=['time','error_message'],header=0)
#df = pd.read_csv('data/error0101.txt')
#df = pd.read_csv('data/error0102.txt')
#df = pd.read_csv('data/error0115.txt')
#df = pd.read_csv('data/error0201.txt')
#df = pd.read_csv('data/error0225.txt')
```

```python
cluster = pipeline.Chain(df, target='error_message', mode='create',matching_accuracy=0.7)
```

```python
cluster.process()
```

```python
#some of the clusters
cluster.result.head(5)
```

```python
#some messages from the cluster 2
print(cluster.in_cluster(cluster.result, 2),"\n")
```

```python
#cluster.df.to_csv("ascii_errors.txt",index=False)
```

```python
cluster.result['pattern'][85]
```

```python
clusters, outliers = cluster.split_clusters(cluster.result, 'cluster_size', threshold=3)
outliers.head(10) #clusters containing fewer than 3 messages
clusters.head(20)
#clusters.to_csv('ascii_patterns.txt',index=False)
```

```python
#top 20 clusters by the size of cluster
top_numb = 20
top_clusters = cluster.validation(clusters)[['cluster_size','pattern','mean_similarity']].sort_values(by=["cluster_size"], ascending=False) #clusters sorted by similarity inside the cluster
top_clusters.head(top_numb)
```

```python
#most inaccurate clusters
top_clusters[['cluster_size','pattern','mean_similarity']].sort_values(by=["mean_similarity"]).head(10)
```

```python
perc = round(top_clusters['cluster_size'].head(20).sum()/(df.shape[0]-1)*100,2)
f"{perc}% of all logs are in the top {top_numb} of {len(cluster.result)} clusters"
```
